## **Complext Vectors** 

---
### **Definition:** A vector contains complex number is a complext vector.

### Example of complext vector in a four dimentional space: 

$$\LARGE \color{blue}{\textbf{v}}  \, 
   = \,\,\
    \color{blue}{\begin{bmatrix}
    1+3i \\
    -2i \\
    4 \\
    -5
    \end{bmatrix}
  }\,\in \mathbb{C}^4
$$

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)
## **Dot Product with Complext Vectors:** Same definition as the dot product with real numbers 

$$\Large
\begin{align}
\langle\textbf{v}, \textbf{w} \rangle  \, 
   &= \,\,
   \begin{bmatrix}
    \color{green}{1}+\color{orange}{3i} \\
    \color{orange}{-2i} \\
    \color{green}{4} \\
    \color{green}{-5}
    \end{bmatrix}^{\textbf{T}}\,
\begin{bmatrix}
    \color{green}{6}+\color{orange}{2i} \\
   \color{green}{7}\\
   \color{orange}{3i} \\
   \color{green}{-2}
    \end{bmatrix} \\ \\
    & =(\color{green}{1}+\color{orange}{3i})(\color{green}{6}+\color{orange}{2i}) +(\color{orange}{-2i} )(\color{green}{7}) +(\color{green}{4})(\color{orange}{3i} ) +(\color{green}{-5})(\color{green}{-2}) \\
    & = \color{green}{1} +\color{orange}{2i}+ \color{orange}{18i}\color{green}{-6} - \color{orange}{14i} + \color{orange}{12i}+\color{green}{10}\\
    & = \color{orange}{4i}+ \color{green}{5}
\end{align}
$$

## **Length of a Complex Vector**

$$\Large \text{length of a complext vector: }\,\, |\mathbf{z}| = \sqrt{\mathbf{z}^{\text{H}}\mathbf{z}}=\sqrt{\mathbf{z}^{\text{*}}\mathbf{z}}$$

### where $$\mathbf{z}^{\text{H}} = \mathbf{z}^{\text{*}} = \text{Hermitian or conugate transpose of vector }\mathbf{z}$$
