### **ðŸ”– $\color{blue}{\textbf{Practice: }}$ Method 1: Using native Python** code to translate the mathematical definition of a dot product of vectors **v1** and **v2**. $$\LARGE\sum_i^n\mathbf{v1}_i\cdot \mathbf{v2}_i$$

```python
# create 2 vectors as list
v1 = [6, 5, 4, 2, 1]
v2 = [0, 6, 4, 6, 5]

# verify if a dot product is valid
if (len(v1) != len(v2)):
    print(f"the dimensionality of the 2 vectors {len(v1)} and {len(v2)} do not match. So no dot product.")
else:
    # intialize a variable to stoe the dot product as a sum
    dot_product = 0
    for i in range(len(v1)): # loop through the length of the vectors
        # this indentation is for the for loop
        dot_product += v1[i] * v2[i] # can never skip a multiplication in any programming language

print(f"If v1 = {v1} and v2 = {v2}")
print(f"Their dot product = {dot_product}")

```

### **ðŸ”– $\color{blue}{\textbf{Practice: }}$ Method 2: [`numpy.multiply()`](https://numpy.org/doc/stable/reference/generated/numpy.multiply.html)** code to translate the mathematical definition of a dot product of vectors **v1** and **v2**. 
$$\LARGE\sum_i^n\mathbf{v1}_i\cdot \mathbf{v2}_i$$

```python
import numpy as np

# create 2 vectors as list
v1 = [6, 5, 4, 2, 1]
v2 = [0, 6, 4, 6, 5]

# verify if a dot product is valid
if (len(v1) != len(v2)):
    print(f"the dimensionality of the 2 vectors {len(v1)} and {len(v2)} do not match. So no dot product.")
else:
    # intialize a variable to stoe the dot product as a sum
    dot_product = 0
    for i in range(len(v1)): # loop through the length of the vectors
        # this indentation is for the for loop
        # dot_product += v1[i] * v2[i] # can never skip a multiplication in any programming language
        dot_product += np.multiply(v1[i], v2[i])


print(f"If v1 = {v1} and v2 = {v2}")
print(f"Their dot product = {dot_product}")

```

### **ðŸ”– $\color{blue}{\textbf{Practice: }}$ Method 3: Using `numpy` library [`numpy.multiply()`](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html)** code to translate the mathematical definition of a dot product of vectors **v1** and **v2**. 
$$\LARGE\mathbf{a\cdot b} \,= \,\langle \mathbf{a},  \mathbf{b}\rangle\, =\, \mathbf{a}^{\text{T}}\mathbf{b} $$

```python
import numpy as np

# create 2 vectors as list
v1 = [6, 5, 4, 2, 1]
v2 = [0, 6, 4, 6, 5]

# verify if a dot product is valid
if (len(v1) != len(v2)):
    print(f"the dimensionality of the 2 vectors {len(v1)} and {len(v2)} do not match. So no dot product.")
else:
    # intialize a variable to stoe the dot product as a sum
    dot_product = sum(np.multiply(v1, v2))


print(f"If v1 = {v1} and v2 = {v2}")
print(f"Their dot product = {dot_product}")

```
