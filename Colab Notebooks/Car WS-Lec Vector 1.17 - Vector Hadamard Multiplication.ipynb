## **Definition of Hadmard (Element-wise) Multiplication** 


###$$
  \Large \color{blue}{\textbf{v}}\odot \color{magenta}{\textbf{w}} \, = \,\,\stackrel{\stackrel{\LARGE \color{blue}{\textbf{v}}}{\mbox{   }}}{%
    \color{blue}{\begin{bmatrix}
    1 \\
    0 \\
    2 \\
    5\\
    -2
    \end{bmatrix}}%
  }\, \odot
  \stackrel{\stackrel{\LARGE \color{magenta}{\textbf{w}}}{\mbox{   }}}{%
    \color{magenta}{\begin{bmatrix}
    2 \\
    8 \\
    -6 \\
    1\\
    0
    \end{bmatrix}}%
  }\ \quad= 
\begin{bmatrix}
\begin{align}
\color{blue}{1} &\cdot \color{magenta}{2}\\
\color{blue}{0}  &\cdot \color{magenta}{8}\\
\color{blue}{2} &\cdot (\color{magenta}{-6})\\
\color{blue}{5} &\cdot \color{magenta}{1}\\
(\color{blue}{-2}) &\cdot \color{magenta}{0}\\
\end{align}
\end{bmatrix}\, = \, 
\begin{bmatrix}
2\\
0\\
-12\\
5\\
0\\
\end{bmatrix}
  $$

## **ðŸ”– $\color{blue}{\textbf{ Practice: }}$Python for Hadmard Multiplication** 

### **1. Vectors are defined $\color{red}{\textbf{list}}$**

```python
import numpy as np

w1 = [1, 3, 5]
w2 = [3, -4, 2]

w3 = np.multiply(w1, w2)
print(f"Hadmard multiplication of two vectors as list")
print(f"vector1 = {w1} and vector2 = {w2}")
print(f"The hadmard multiplication of vector1 and vector2 = {w3}")
```

### **2. Vectors are defined as $\color{red}{\textbf{arrays}}$**

```python
import numpy as np

w1 = [1, 3, 5]
w2 = [3, -4, 2]

w3 = np.array(w1) * np.array(w2)
print(f"Hadmard multiplication of two vectors as numpy array")
print(f"vector1 = {w1} and vector2 = {w2}")
print(f"The hadmard multiplication of vector1 and vector2 = {w3}")
```

####**Not everyone** uses the same notation conventions all the time. so it's important for you to learn to be a little bit **flexible** and to see the math and the concepts, even if the notations are slightly different.
