## **Vector Length: Algebra** 

$$\Large\left\Vert  \mathbf{v} \right\Vert \, =\, \sqrt{\mathbf{v}^{\text{T}}\mathbf{v}}\, = \, \sqrt{\sum_i^n\mathbf{v}_i\mathbf{v}_i}$$ 
####where **v** is a vector.

![green-divider](https://user-images.githubusercontent.com/7065401/52071924-c003ad80-2562-11e9-8297-1c6595f8a7ff.png)

## **2d Vector Length and Pythogorean Theorem** 

$$\Large \text{Given a vector }\mathbf{v} = \color{blue}{\begin{bmatrix}
    \mathbf{v}_1 \\
    \mathbf{v}_2  
\end{bmatrix}} \quad \text{norm: }\left\Vert  \mathbf{v} \right\Vert = \sqrt{\color{blue}{\mathbf{v}_1}^2+\color{blue}{\mathbf{v}_2}^2}$$

````python
import numpy as np

# create 2 vectors as list
v1 = [6, 5, 0.2, 321, 0.1, 3.2, 8]
# v2 = [0, 6, 4, 6, 5]

# 1. copy the code from the above method 1 to 5
# 2. calculate the square root of the result, np.sqrt(a), a ** (1/2), a ** (0.5)
# save the norm for each method in v_len_1 ... to v_len_5

# Method 1
dot_product_1 = 0
for i in range(len(v1)):
    dot_product_1 += v1[i] * v1[i]

# method 2
dot_product_2 = 0
for i in range(len(v1)):
    dot_product_2 += np.multiply(v1[i], v1[i])

# method 3
dot_product_3 = sum(np.multiply(v1, v1))

# method 4
dot_product_4 = np.matmul(v1, v1)

# method 5
dot_product_5 = np.dot(v1, v1)

v_len_1 = np.sqrt(dot_product_1)
v_len_2 = np.sqrt(dot_product_2)
v_len_3 = np.sqrt(dot_product_3)
v_len_4 = np.sqrt(dot_product_4)
v_len_5 = np.sqrt(dot_product_5)

print(f"dot_product_1: {dot_product_1}, v_len_1: {v_len_1}\n")
print(f"dot_product_2: {dot_product_2}, v_len_2: {v_len_2}\n")
print(f"dot_product_3: {dot_product_3}, v_len_3: {v_len_3}\n")
print(f"dot_product_4: {dot_product_4}, v_len_4: {v_len_4}\n")
print(f"dot_product_5: {dot_product_5}, v_len_5: {v_len_5}\n")
````
