## **Sign and Orthogonality of Dot Product of Two Vectors Based on the Angle Between Them** 

<details>
  <summary><b>Show Dot Product Features</b></summary>

<p style="text-align:center;"><img src = "https://drive.google.com/uc?id=1AKdnE-n827yMlUDnvtVuabjyxLhdV0Z0" width = "1400px" height = "500px" />

* The **cosine** of an angle is bound between minus one and plus one $-1\le \theta \le 1$.

* The **lengths of vectors** also are strictly non-negative numbers $|\mathbf{a}| \ge 0$

* The combination of the range of $\theta$ with the non-negative lengths of vectors gives us some interesting insights into the DOT product and how the sign of the DOT product is related to the relationship between the two vectors.

That's because when you look at the formula $$ \langle \mathbf{a, b} \rangle= |\mathbf{a}| |\mathbf{b}|\text{cos}( \theta_{ab} $$ here, you see that the two factors $|\mathbf{a}|$ and  $|\mathbf{b}|$ are always going to be **non-negative.**  

* Unless one of the is zero vector or $|\mathbf{a}|$ and  $|\mathbf{b}|$ will be positive.  So the the only factor in this term that can possibly be negative is the **cosine** of the angle between the two vectors.

* If the cosine of the angle between the two vectors is less than $90^âˆ˜$ degrees or $\pi/2$ then you would say that the two vectors meet at an acute angle.

  And the cosine of that angle is necessarily going to be positive because the cosine of an acute angle is positive. So that means that the DOT product must be positive.

* If you see a DOT product between two vectors that is greater than zero, then that indicates that those two vectors have an angle between them of less than $90^âˆ˜$.

* If the angle is obtuse or if it's greater than $90^âˆ˜$ degrees, then the cosine will be negative. And that means that the DOT product will also be negative.

* If the angle between the two vectors is exactly $90^âˆ˜$ degrees or $\pi/2$ making this a right triangle outside a right angle. Then the DOT product is zero.

  And that's because the cosine of 90 degrees of the cos$(\pi/2) = 0$.  So looking back at this term, you have zero times something time something. So it doesn't actually matter how **long** or how **short** these two vectors are.

* If they meet at a right angle, then the DOT product between these two vectors is necessarily going to be zero.

  This is such an important case that is given its own special name of orthogonal. And **orthogonal** is an important word, it's an important concept that will come up very often in linear algebra.

* When the two vectors are pointed in the opposite direction ($\theta$ is 180 degrees or $\pi$ radians).

  Then vector **b**, for example, would be pointing exactly this other direction as vector **a**, as if you would multiply this vector by a scalar minus one.

  In that case, the cosine of 180 degrees or $\pi$ is minus one.  So then the DOT product would be basically the same thing.  But with a negative sign in front of it.

* Finally when the vectors point in exactly the same direction, then the angle between them is zero.  So in this case, the angle between the two vectors is zero and the cosine of zero is one.

  So that means in this case, the DOT product actually reduces to the multiplication of the length of the two vectors.

* So **b** is equal to **a**, then the dot product of the vector with itself is simply its length times its length.

  And that's the length squared, of course and the square root of that is the length that provides an additional geometric explanation for why the length of a vector is computed as the square root of the dot product of the vector with itself.
</details>

## **Python for visualization of dot product**

### **ðŸ”– $\color{blue}{\textbf{Practice: }}$ Compute the numerical value of the angle using Python**

```python
import numpy as np
import math

v1 = np.array([2, 4, -3]) #numpy does not specify is a vector is a row or col
v2 = np.array([0, -3, -3])

# compute the radian between v1 and v2

ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

print(f"the angle between {v1} and {v2} is {ang: .1f} in radian and {ang/np.pi * 180: .2f} in degrees")
```

## **ðŸ”– $\color{blue}{\textbf{Practice: }}$Experical Proof of Algebraic and Geometric Formulas of Dot Product**

```python
# prove equvilancy of algebraic and geometric dot product formulas

import numpy as np
import math

v1 = np.array([2, 4, -3]) #numpy does not specify is a vector is a row or col
v2 = np.array([0, -3, -3])

# compute the radian between v1 and v2

# dot product algebraic way
dp_a = np.dot(v1, v2)

# Geometric way
ang = np.arccos(np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2)))

# dot product
dp_g = np.linalg.norm(v1) * np.linalg.norm(v2) * np.cos(ang)

print(f"the angle between {v1} and {v2} is {ang: .1f} in radian and {ang/np.pi * 180: .2f} in degrees")
print(f"dot product using algebra = {dp_a}")
print(f"dot product using geometry = {dp_g}")
```
