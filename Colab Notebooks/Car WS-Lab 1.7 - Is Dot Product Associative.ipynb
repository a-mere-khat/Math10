#### **3: Special Case** All vectors with the same dimention and one of the vector is zero.:

```python
#@title All same dimension, one vector is 0
import numpy as np
# 1. define 3 dimensional variables
n = 5
m = 5
k = 5

count_false = 0

for i in range(10000):
    # create 3 random vectors corresponding to the 3 dimensions
    a = np.zeros(n)         # a = [0] * n
    b = np.random.randn(m)
    c = np.random.randn(k)

    # calculate the left and right side of the associative property
    left = np.dot(a, np.dot(b, c))
    right = np.dot(np.dot(a, b), c)
    if np.array_equal(left, right):
        count_false += 1

# compare the answers
print(f"vector a = {a}")
print(f"vector b = {b}")
print(f"vector c = {c}")
print(f"\n left and right sides of the associative property formula")
print(f" left = {left}")
print(f"right = {right}")
print(f"\n difference between the left and the right side of the associative property\n{left - right}")

print(f"The number of left and right side are equal in the associative property is {count_false}")
```

##### **4: Scecial Case** all vectors are equal

```python
#@title All vectors are equal
import numpy as np
# 1. define 3 dimensional variables
n = 5
m = 5
k = 5

count_false = 0

for i in range(10000):
    # create 3 random vectors corresponding to the 3 dimensions
    b = np.random.randn(m)
    a = c = b

    # calculate the left and right side of the associative property
    left = np.dot(a, np.dot(b, c))
    right = np.dot(np.dot(a, b), c)
    if np.array_equal(left, right):
        count_false += 1

# compare the answers
print(f"vector a = {a}")
print(f"vector b = {b}")
print(f"vector c = {c}")
print(f"\n left and right sides of the associative property formula")
print(f" left = {left}")
print(f"right = {right}")
print(f"\n difference between the left and the right side of the associative property\n{left - right}")

print(f"The number of left and right side are equal in the associative property is {count_false}")
```
