### **$\color{brown}{\textbf{TODO 1: }}$ State the Inequality**



The Cauchyâ€“Schwarz inequality states that for all vectors $\mathbf{u}$ and $\mathbf{v}$ of an inner (dot in $R^2$ or $R^3$ product space


$$ | \mathbf{u}^T \mathbf{v} | \leq || \mathbf{u} ||\, ||\mathbf{v} || \, \textbf{(Cauchy-Schwarz inequality [written using only the inner product])}$$

where $\mathbf{u}^T \mathbf{v}  $  is the inner product. The two bars $| \, |$ of vertical lines here indicates the absolute value.  Because the DOT product is just one number. So we just need the absolute value.

And here the double vertical lines $||\, \, ||$ refer to the vector magnitude or the length of these two vectors $\mathbf{u}$ and $ \mathbf{v}$.

So the Couchy Schwartz inequality says that the absolute value or the magnitude of the DOT product between two vectors $\mathbf{u}$ and $ \mathbf{v}$ is less than or equal to the product of the norms of those two vectors

Examples of inner products include the real and complex dot product; see the examples in inner product. Every inner product gives rise to a *norm*, called the canonical or *induced norm*, where the norm of a vector $ \mathbf{u} $  is denoted and defined by:


$$ \|\mathbf {u} \|:={\sqrt {\mathbf{u}^T \mathbf{u} }}$$


so that this norm and the inner product are related by the defining condition $\|\mathbf {u} \|^{2}=|\mathbf{u}^T \mathbf{u}| ,$ where $|\mathbf{u}^T \mathbf{u} |$ is always a non-negative real number (even if the inner product is complex-valued). By taking the square root of both sides of the above inequality, the Cauchyâ€“Schwarz inequality can be written in its more familiar form:

$$ |\mathbf{u}^T \mathbf{v} |\leq \|\mathbf {u} \|\|\mathbf {v} \|.\, \textbf{(Cauchy-Schwarz inequality [written using only the inner product])}$$


Moreover, $\color{red}{\text{the two sides are equal if and only if $ \mathbf {u} $ and $ \mathbf {v} $ are linearly dependent}}$.
