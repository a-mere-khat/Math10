## **Is the dot product distributive** 


<details>
  
<summary><b>Formal Mathematical definition </b></summary>

**Definition of distributive of multiplication over additions by Wiki**

Given a set $S$ and two binary operation * and $+$ on $S$,
* the operation * is *left-distributive over * (or with respect to) $+$ if, given elements $x, y, $ and $z$ of $S$,
$$x * (y+z) = (x*y) + (x*z);$$
* the operation * is *right-distributive* over $+$ if, given any elements $x, y$, and $z$ of $S$,
$$(y+z)*z=(y*x)+(z*x);$$
* and the operation * is *distributive* over $+$ if it is left- and right-distributive.

When * is commutative, the three conditions above are logically equivalent.

</details>

## **Is dot product associative?** 


<details>
  
<summary><b>Formal Mathematical definition </b></summary>

Finally, a binary operation * on a set $S$ is called **associative** if it satisfies the **associative law**:

$$(x * y) * z = z * ( y * z) $$ for all $x, y, z$, in $S$

## **Is the dot product commutative** 


<details>
  
<summary><b>Formal Mathematical definition </b></summary>

**Definition of distributive of multiplication over additions by [Wiki](https://en.wikipedia.org/wiki/Commutative_property)**

Finally, a binary operation * on a set $S$ is called **commutative** if :

$$x * y = y * x $$ for all $x, y$, in $S$

An operation that does not satisfy the above property is called non-commutative.
