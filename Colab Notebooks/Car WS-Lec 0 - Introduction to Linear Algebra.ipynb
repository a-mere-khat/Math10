## **Vectors and Matrices**

### **1. A vector is an ordered list of numbers:**

* A vector is an ordered list of numbers that can be arranged in a column like this, or arranged in a row.

* Vectors can also represent functions, for example a function of time. 


$$
  \stackrel{\stackrel{\mbox{Column Vector}}{\mbox{   }}}{%
    \begin{bmatrix}
    -1 \\
    2 \\
    3 + 6i \\
    0.22\\
    \frac{1}{9}
    \end{bmatrix}%
  }\ \quad
  \stackrel{\stackrel{\mbox{Row Vector}}{\mbox{   }}}{%
    \begin{bmatrix}
    1 & a & 3 & \\ 
    \end{bmatrix}%
  }\ \quad
  \stackrel{\stackrel{\mbox{Vector}}{\mbox{ as a Function of time }}}{%
    \begin{bmatrix}
    sin(t) \\
    t \\
    cos(t) \\
    \end{bmatrix}%
   }
$$

### **2. A matrix is a spreadsheet of numbers:**

$$
\stackrel{\stackrel{\mbox{Matrix is a Spread Sheet}}{\mbox{ }}}{%
    \begin{bmatrix}
    a_{11} & a_{12} & \cdots & a_{1M} \\
    a_{21} & a_{22} & \cdots & a_{2M} \\
    \vdots & \vdots & \ddots & \vdots \\
    a_{N1} & a_{N2} & \cdots & a_{NM}
    \end{bmatrix}%
  }\ \quad
$$

### **3.** Transformation and Rotations:

The ease and power of transformation matrices is also why linear algebra is important for computer graphics, graphical programming, and for geometry.  Matrices can be used to encode translations, rotations, or stretching or compressing factors.  So it's really easy to take an image and compress it, expand it, or rotate it along some axis.


```python
#@title Import and display image from my google drive

#@title Mount my Google Drive
from google.colab import drive

drive.mount("/content/gdrive")

# variable for path to the imge file
img_path = "/content/gdrive/MyDrive/MATH 010 Linear Algebra - Sp 23/Lectures/0. Introduction to Linear Algebra/Einstein_tongue.jpg"

img = Image.open(img_path)

from PIL import Image
import PIL


# rotating the image 60 deg counter clockwise
# notice you cannot rotate the image in numpy array 
img_rotated = img.rotate(60, PIL.Image.NEAREST, expand = 1)


import matplotlib.pyplot as plt


#@title Image Rotation by Matrix
plt.figure(figsize=(6 * 2, 4 * 1))

# subplots (numbers of rows, numbers of columns, ith plots)

plt.subplot(1, 2,1)
plt.imshow(img,  cmap = "gray")
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
 
plt.title("Original Image")

plt.subplot(1, 2,2)
plt.imshow(img_rotated,  cmap = "gray")
plt.tick_params(left = False, right = False , labelleft = False ,
                labelbottom = False, bottom = False)
 
plt.title("Image Roated by $60^\circ$")

plt.show()
```

Here you see an example.  This is a picture of Eistain.

The mechanism of this rotation is by transforming all the coordinates in the original picture to new coordinates, and those new coordinates are defined simply by multiplying the $(x y)$ of the original picture by this particular matrix.

$$
    \begin{bmatrix}
    cos(\theta) & -sin(\theta)  \\
    sin(\theta) & cos(\theta)
    \end{bmatrix} \hspace{20mm}\theta = \pi/3
$$

This is also called a transformation matrix and it simply rotates the image coordinates as you see here.

In this particular example I used an angle of theta value of pi over 3, which is 60 degrees.

### **4. Describing Objects** There is a direct link between linear algebra and geometry, and matrices can be interpreted as descriptions of geometric objects like lines or conics for example.

Here you see an example of a beautiful surface that represents the information in a two by two matrix.

The function that defines this surface is called the normalized quadratic form. It's used to extract key features in a matrix in statistics and in data compression, and of course you will learn all about this particular operation later on in the course.

```python
#@title Quadratic Surface from $2\times 2$ matrix
import numpy as np

# Visualize raw and normalized quadratic form

A = np.array([[-2, 3], [2, 8]])

n = 30

xi = np.linspace(-2, 2, n)

# for the visuzalization
X, Y = np.meshgrid(xi, xi)

# initialize data set
qf = np.zeros((n, n)) # quadratic form
qfN = np.zeros((n,n)) # normalized quadratic form

for i in range(n):
    for j in range(n):
        # create x coordinators a column vector
        x = np.transpose([xi[i], xi[j]])
        # compute the quadratic form
        qf[i, j] = x.T @ A @ x
        qfN[i, j] = qf[i, j] / (x.T @ x)

import plotly.graph_objects as go

fig= go.Figure()

fig.add_trace(go.Surface(
                          x = X, y = Y, z = qf,colorscale = "Rainbow", showscale=False
                         
                        ))
fig.update_traces(contours_z=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_z=True))
fig.update_traces(contours_y=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_y=True))
fig.update_traces(contours_x=dict(show=True, usecolormap=True,
                                  highlightcolor="limegreen", project_x=True))

axis = dict(
showbackground=True,
#backgroundcolor="white",
backgroundcolor="rgba(0,0,0, 0.1)",
showgrid=False,
zeroline=False,
showline=False,
showticklabels = False,
title = ""
)
ztickvals=list(range(-3,4))
fig.update_layout(title='Quadratic Form Energy', autosize=False,
                  scene_camera_eye=dict(x=1.87, y=0.88, z=-0.64),
                  width=800, height=800,
                  margin=dict(l=65, r=50, b=65, t=90),
                  scene=dict(xaxis=dict(axis, range=[-3.5, 3.5]),
                            yaxis=dict(axis, range=[-3.5, 3.5]),
                            zaxis=dict(axis , tickvals=ztickvals),
                            #aspectratio=dict(x=1,
                            #                 y=1,
                            #                 z=0.95)
                           ),

                  )
fig.show()
```

### **4. Singular Value Decomposition of the Image** 
If this looks weird and mysterious and threatening and intimidating, don't worry.  You are going to understand all of what this line of code is actually implementing, what these different matrices mean towards the end of this course.

---
##### **a. SVD:** Briefly, the goal of singular value decomposition is to take a **matrix** and decompose that matrix into **three other matrices**.
####$$\text{A = U}\Sigma\text{V}^\text{T}$$

####in code `U, S, V = numpy.linalg.svd(A)`

* The three matrices U, $\Sigma$ and V are basically telling us about important **spatial patterns** in this image or more generally, important **directions in the space** in which this matrix lives.

* Because this matrix is a **picture,** then this is going to find **patterns** in the image that are **important**.


```python
U, S, V = np.linalg.svd(pic)
```

```python
import plotly.graph_objects as go
import numpy as np

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = np.arange(len(S)), y = S,
        mode = "markers+lines",
        marker_size = 6,
        line_color = "blue"
    )
)

fig.show()
```
