## **Vector-Scalar Multiplication: Geometry** 




```python
#@title Gemetry of Vector-scalar multiplication with $\mathbf{W}=\lambda \mathbf{v}$
import numpy  as np
import matplotlib.pyplot as plt

# define two vectors 

v = np.array([1, 1])
v0 = [0, 0]

def draw(vec_tail, vec_head, color, width):
    """
    Draw vectors
    inputs: vectors for tail and head, and color of the vector
    """
    array = np.array([[vec_tail[0], vec_tail[1], vec_head[0], vec_head[1]]])
    X, Y, U, V = zip(*array)
    plt.quiver(X, Y, U, V,color=color, width=width, headwidth=3., headlength=4, angles='xy', scale_units='xy',scale=1)
    #ax.quiver(X, Y, U, V,color=color)

# set the common features of all subplots

# dynamically set the axis limits
lim_low = -1
lim_high = 2.5

# Create of figure with subplots of 5x5 each in 1 row and 3 column 
plt.figure(figsize=(5 * 3, 5 * 1))  # width by height

##. first subplot for lambda > 1

l = 1.8      # A Scalar
lv = l * v   # Vecor-Scalar Multiplication

# plot the vectors in row 1 column 1
plt.subplot(1, 3,1)
ax = plt.gca()


# Add vectors by scall the function draw given the coordinate and tail and head and color
draw([0, 0], v, "b", 0.015)
plt.plot([v0[0], lv[0]], [v0[1], lv[1]], 'r--', linewidth=3, markersize=12)
draw(l*0.9 * v , l *0.1 * v, "r", 0.02)
# Set ploting limits
# make sure that it covers the range of all vectors

plt.axis("square")
plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

#plt.tick_params(left = standard, right = standard , labelleft = standard ,
#                labelbottom = standard, bottom = False)
#plt.figure().patch.set_visible(standard)
#if standard == False:
#  ax.axis('off')
plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.grid()
plt.title(f"$\lambda$v \n$\lambda > 1$")
plt.legend([r"$\lambda \mathbf{v}$", "$\mathbf{v}$"])

### second subplot for 0< lambda < 1
l = 0.7
lv = l * v

# plot the vectors in row 1 column 1
plt.subplot(1, 3,2)
ax = plt.gca()


# Add vectors by scall the function draw given the coordinate and tail and head and color
draw([0, 0], v, "b", 0.015)
plt.plot([v0[0], lv[0]], [v0[1], lv[1]], 'r--', linewidth=3, markersize=12)
draw(l*0.8 * v , l *0.2 * v, "r", 0.04)
# Set ploting limits
# make sure that it covers the range of all vectors

plt.axis("square")
plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.tick_params(left = False, right = False , labelleft = False , labelbottom = True, bottom = False)
#plt.figure().patch.set_visible(standard)
#if standard == False:
#  ax.axis('off')
#plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.grid()
plt.title(f"$\lambda$v \n$\lambda \in(0,1)$")
plt.legend([r"$\lambda \mathbf{v}$", "$\mathbf{v}$"])

### third column subplot for lambda < 0
l = -0.8
lv = l * v

# plot the vectors in row 1 column 3
plt.subplot(1, 3,3)
ax = plt.gca()


# Add vectors by scall the function draw given the coordinate and tail and head and color
draw([0, 0], v, "b", 0.015)
plt.plot([v0[0], lv[0]], [v0[1], lv[1]], 'r--', linewidth=3, markersize=12)
draw(l*0.8 * v , l *0.2 * v, "r", 0.04)
# Set ploting limits
# make sure that it covers the range of all vectors
plt.axis("square")
plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.tick_params(left = False, right = False , labelleft = False , labelbottom = True, bottom = False)
#plt.figure().patch.set_visible(False)
#if standard == False:
#  ax.axis('off')
#plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.grid()
plt.title(f"$\lambda$v \n$\lambda < 0$")
plt.legend([r"$\lambda \mathbf{v}$", "$\mathbf{v}$"])
plt.show()
```
