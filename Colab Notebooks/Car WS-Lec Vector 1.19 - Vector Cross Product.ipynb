## **Algebraic Notation and Definition of a cross product** 
$$\Large
\begin{align}\text{Cross product of vectors } \color{blue}{\mathbf{v}} \text{ and }\color{magenta}{\mathbf{w}}\,\,  &= \,\,\color{blue}{\mathbf{v}}\times\color{magenta}{\mathbf{w}}\\
\text{Cross product of vectors } \color{magenta}{\mathbf{w}}\text{ and }\color{blue}{\mathbf{v}}\,\,  &= \,\,\color{magenta}{\mathbf{w}}\times\color{blue}{\mathbf{v}} =  \color{orange}{-} \,\,\color{blue}{\mathbf{v}}\times\color{magenta}{\mathbf{w}}
\end{align}$$

Equation of Planes

Suppose that we have given three points $\mathbf{r}_0$, $\mathbf{r}_1$, and $\mathbf{r}_2$, that are not co-linear.  Then, these points define a plane, and the vector $\mathbf{v}_1 =\mathbf{r}_1 -\mathbf{r}_0$ and $\mathbf{v}_2 =\mathbf{r}_2 -\mathbf{r}_0$ are vectors contained **within the plane**, that are also not parallel to one another.

A plane consists of all vectors that are orthogonal to a given direction $\mathbf{n}$, which is said to be *normal* to the plane and passes through a given points $\mathbf{r}_0$.  The *normal vector* $\mathbf{n}$ can be obtained by computing
$$\mathbf{n} = \mathbf{v}_1 \times \mathbf{v}_2.$$

Let $\mathbf{r}$ be *any point* in the plane.  Then the vector $\mathbf{u}=\mathbf{r} -\mathbf{r}_0$ is orthogonal to $\mathbf{n}$.  That is, 
$$\mathbf{n}\cdot  \mathbf{u}= \mathbf{n}\cdot(\mathbf{r} - \mathbf{r}_0)=0.$$

This equation is called the *vector equation of the plane*.  

If we write 
$$\mathbf{n} = \langle a, b, c \rangle, \quad\mathbf{r} = \langle x, y, z \rangle, \quad\mathbf{r}_0 = \langle x_0, y_0, z_0 \rangle$$ then the vector equation can be rewritten as 
$$ax+by+cz+d=0$$
where  $d =- \mathbf{n}\cdot \mathbf{r}_0 = -(ax_0 + by_0 + cz_0)$.  This is a *linear equation* in the *unknowns* $z, y,$ and $z$.

### **Plot Cross Product and Plane in Python**

#### **Static Graph with `matplotlib.pyplot`**

```python
#@title
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# create vectors
v1  = [ -3,  2, 5 ]
v2  = [  4, -3, 0 ]

# Python's cross-product function

v3a = np.cross( v1,v2 )


fig = plt.figure()
ax = fig.gca(projection='3d')

# draw plane defined by span of v1 and v2
xx, yy = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]
ax.plot_surface(xx,yy,z1,alpha=.2)

## plot the two vectors
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'k')
ax.plot([0, v2[0]],[0, v2[1]],[0, v2[2]],'k')
ax.plot([0, v3a[0]],[0, v3a[1]],[0, v3a[2]],'r')


ax.view_init(azim=150,elev=45)
plt.show()
```

```python
#@title Visualization of Vectors { run: "auto" }
#@markdown Enter the three coordinate x1, y1, z1 of the vector v1
x1 = -3 #@param {type:"number"}
y1 = 2 #@param {type:"number"}
z1 = 5 #@param {type:"number"}

#@markdown Enter the three coordinate x2, y2, z2 of the vector v2
x2 = 4 #@param {type:"number"}
y2 = -3 #@param {type:"number"}
z2 = 0 #@param {type:"number"}

import numpy as np
import plotly.graph_objects as go


# create vectors
v1 = np.array([ x1,  y1, z1 ])
v2 = np.array([ x2,  y2, z2 ])

# Python's cross-product function

v3a = np.cross( v1,v2 )
mesh_size = 0.1

#axis_lim = abs(max(abs(Vs))

x_range = max(max(abs(v1)), max(abs(v2))) * 2


xrange = np.arange(-x_range, x_range, mesh_size)
yrange = np.arange(-x_range, x_range, mesh_size)
xx, yy = np.meshgrid(xrange, yrange)

# draw plane defined by span of v1 and v2
#xx, yy = np.meshgrid(np.linspace(-10,10,10),np.linspace(-10,10,10))
z1 = (-v3a[0]*xx - v3a[1]*yy)/v3a[2]

lim_fig =  max(max(abs(v1)), max(abs(v2)), max(abs(v3a))) +1

axis = dict(showgrid=True, zeroline=True, zerolinecolor = "black", showticklabels=True, nticks = 2,
                       range = [-lim_fig, lim_fig]
            )


fig = go.Figure()

fig.add_traces(go.Surface(x=xrange, y=yrange, z=z1, name='pred_surface', colorscale = "peach", opacity = 0.5, showscale = False))
#fig.update_traces(marker=dict(size=5))
# Plot the two vectors for cross product


# plot v1
fig.add_trace(go.Scatter3d(x=[0, v1[0]], y=[0, v1[1]], z = [0, v1[2]], 
                         mode = "lines", line = dict(color = 'green', width= 10),
                         ))

fig.add_trace(
              go.Cone(
                      x=[v1[0]],
                      y=[v1[1]],
                      z=[v1[2]],
                      u=[v1[0]],
                      v=[v1[1]],
                      w=[v1[2]],
                    sizemode="absolute",
                    sizeref=1,
                    anchor="tip",
                    colorscale='greens',
                    showscale=False,
    ))

#plot v2
fig.add_trace(go.Scatter3d(x=[0, v2[0]], y=[0, v2[1]], z = [0, v2[2]], 
                         mode = "lines", line = dict(color = 'blue', width= 10),
                         ))

fig.add_trace(
              go.Cone(
                      x=[v2[0]],
                      y=[v2[1]],
                      z=[v2[2]],
                      u=[v2[0]],
                      v=[v2[1]],
                      w=[v2[2]],
                    sizemode="absolute",
                    sizeref=1,
                    anchor="tip",
                    colorscale='blues',
                    showscale=False,
    ))



#plot the line
fig.add_trace(go.Scatter3d(x=[0, v3a[0]], y=[0, v3a[1]], z = [0, v3a[2]], 
                         mode = "lines", line = dict(color = 'purple', width= 10),
                         ))

# add the arrow for the cross product
fig.add_trace(
              go.Cone(
                      x=[v3a[0]],
                      y=[v3a[1]],
                      z=[v3a[2]],
                      u=[v3a[0]],
                      v=[v3a[1]],
                      w=[v3a[2]],
                    sizemode="absolute",
                    sizeref=1,
                    anchor="tip",
                    colorscale='purpor',
                    showscale=False,
    ))

fig.update_layout( width = 800,
                title = f'Cross Product  {v1} x {v2}',
                title_x = 0.5,
                margin=dict(r=0, l=0, b=0, t=30),
                scene_aspectmode='cube',
                scene = dict(
                              #xaxis = dict(showgrid=False, zeroline=False, zerolinecolor = "black", nticks=4, range=[-1,4],),
                              #yaxis = dict(showgrid=False, zeroline=False, zerolinecolor = "black", nticks=4, 
                                           #range=[-1,4],),
                              #zaxis = dict(nticks=4, range=[-1,4],),
                              xaxis = axis,
                              yaxis = axis,
                              zaxis = axis,
                              annotations=[
                                          dict(
                                              showarrow=False,
                                              x= v1[0],
                                              y= v1[1],
                                              z= v1[2],
                                              text=f"v1",
                                              xanchor="left",
                                              xshift=10,
                                              opacity=1
                                              ),
                                           dict(
                                              showarrow=False,
                                              x= v2[0],
                                              y= v2[1],
                                              z= v2[2],
                                              #text=f"[{v2[0]}, {v2[1]}, {v2[2]}]",
                                              text=f"v2",
                                              xanchor="left",
                                              xshift=10,
                                              opacity=1
                                              ),
                                             dict(
                                              showarrow=False,
                                              x= v3a[0],
                                              y= v3a[1],
                                              z= v3a[2],
                                              #text=f"[{v2[0]}, {v2[1]}, {v2[2]}]",
                                              text=f"Cross product v1 x v2",
                                              xanchor="left",
                                              xshift=10,
                                              opacity=1
                                              ),
                                          ]
                              ),
                showlegend = False,
                font=dict(
                color="black",
                size=14
            ),
)
# hiding color-bar
fig.show()
```
