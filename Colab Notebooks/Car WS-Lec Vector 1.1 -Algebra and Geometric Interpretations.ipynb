### ðŸ”– **Create a Vector Plotter with `plotly`**

```python
#@title 3d- Vector Plotter
#@markdown Enter the $x, y,$ and $z$ coordinates of the head
x1 = -1.0 #@param {type:"number"}
y1 = 2.0 #@param {type:"number"}
z1 = 3.0 #@param {type:"number"}

import plotly.graph_objects as go
import numpy as np
#axis_lim = abs(max(abs(Vs))
lim_min = min(np.array([x1, y1, z1]))
lim_max = max([x1, y1, z1])
axis = dict(showgrid=True, zeroline=True, zerolinecolor = "black", showticklabels=True, 
            #nticks = 1,
                       #range = [min(min(Vs), min(Vir)) - 1, max(max(Vs), max(Vir)) + 1])
                       #range = [lim_min-1, lim_max+1]
                      )

fig = go.Figure()
fig.add_trace(go.Scatter3d(x=[0, x1], y=[0, y1], z = [0, z1], 
                         mode = "lines", line = dict(color = 'green', width= 6),
                         ))

fig.add_trace(
              go.Cone(
                      x=[x1],
                      y=[y1],
                      z=[z1],
                      u=[x1],
                      v=[y1],
                      w=[z1],
                    sizemode="absolute",
                    sizeref=0.3,
                    anchor="tip",
                    colorscale='greens',
                    showscale=False,
    ))

fig.update_layout( width = 800,
                title = "A Standard Vector",
                title_x = 0.5,
                margin=dict(r=0, l=0, b=0, t=30),
                scene_aspectmode='cube',
                scene = dict(
                              #xaxis = dict(showgrid=False, zeroline=False, zerolinecolor = "black", nticks=4, range=[-1,4],),
                              #yaxis = dict(showgrid=False, zeroline=False, zerolinecolor = "black", nticks=4, 
                                           #range=[-1,4],),
                              #zaxis = dict(nticks=4, range=[-1,4],),
                              xaxis = axis,
                              yaxis = axis,
                              zaxis = axis,
                              annotations=[
                                          dict(
                                              showarrow=False,
                                              x= x1,
                                              y= y1,
                                              z= z1,
                                              text=f"[{x1}, {y1}, {z1}]",
                                              xanchor="left",
                                              xshift=10,
                                              opacity=1),
                                           
                                          ]
                              ),
                showlegend = False,
                font=dict(
                color="black",
                size=14
            ),
)
# hiding color-bar
fig.show()
```

### **Vector Functions**

```python
#@title Illustration of a vector function $ \begin{bmatrix}sin(t) \\tcos(t) \\t \end{bmatrix} \quad$
import plotly.graph_objects as go
import numpy as np

# Helix equation
t = np.linspace(0, 20 * np.pi, 1200)
x, y, z = np.cos(t), t * np.sin(t), t

fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=2,
        color=z,                # set color to an array/list of desired values
        colorscale='Viridis',   # choose a colorscale
        opacity=0.8,
        #showscale = True,
    )
))

# tight layout
layout = go.Layout(
             title = "A Vector Function ",
                  scene = dict(
                    xaxis_title = "sin(t)",
                    yaxis_title = "tcos(t)",
                    zaxis_title = "t"
                  ),
                  margin=dict(l=0, r=0, b=0, t=0))
fig.update_layout(layout)
fig.show()
```

### **List and Array**

#### Sometimes we want to place items that are **similar** in a variable container.  In Python, this type of container is called $\color{blue}{\text{array}}$.  

#### Othertimes, we want to place items that are **NOT similar** in a variable container.  In Python, this type of container is called $\color{blue}{\text{list}}$.

For example, we want to store students grades for several assignemnts.  In this case, we could create a container with student's name and place the score from each assignment in the container.  On the other hand, if we need to store data of a student's ID, majors, enrolled courses, payments, majors, and years in college.  In this case, the container with student's name will a combined data types of numbers and strings in a list.

###ðŸ§š **Common features of a List and Array**

Both the list and array data structures are enclosed in **square brackets "[ ]"**, and allow indexing, slicing, and iterating.

###ðŸ§š **Array vs. List**

<details>
  <summary><b>Comparison Table</b> </summary>

|List |Array |
|--|--|
|Contains elements of different data types|Contains elements of same data types|
|Do not need to be declared| Need to be declared|
|Explicitly importing module is not required to declare a list|Need to import the module explicitly to declare an array|
|Cannot handle arithmetic operations|Can handle arithmetic operations|
|Can be nested inside another list|Must contain all elements of the same size|
|Mostly used in the shorter sequence of data elements|Mostly used in the longer sequence of data elements|
|Easy modifications like addition, deletion, and update of <br>data elements are done|It is difficult to modify an array since addition, deletion <br>and update operation is performed on a single element at a time|
|We can print the entire list without the help of an explicit loop|To print or access array elements, we will require an explicit loop|
|For easy addition of element, large memory storage is required|In comparison to the list, it is more compact in-memory size|

</details>

###ðŸ§š **When should you use list and when should you use array**


<details>
  <summary><b>Show Usages</b> </summary>

* If you need to store a relatively short sequence of items and you **don't plan to do any mathematical operations** with it, a list is the preferred choice. This data structure will allow you to store an ordered, mutable, and indexed sequence of items without importing any additional modules or packages.

* If you have a very **long sequence** of items, consider using an array. This structure offers more efficient data storage.

* If you plan to do any **numerical operations** with your combination of items, use an array. Data analytics and data science rely heavily on (mostly NumPy) arrays.

</details>

#### **Define Vectors**

```python
# 2D vector: row or column
v2_1s = [3, -2] # a list
print(f"2d vector as a list {v2_1s}")

# 2D vector: row or column
v3_1s = [3, -2, 2j] # a list
print(f"3d vector as a list {v3_1s}")
```

```python
# use numpy to create the same 
import numpy as np

# 2D vector: row or column
v2_arr = np.array([[3, -2]]) # a row vector as an array, note double square bracket
print(f"2d vector as an array {v2_arr}")

# option to convert list to array
v2_arr_v1 = np.array([v2_1s])
print(f"2d vector as an array {v2_arr_v1}")

# 2D vector: row or column
v3_arr = np.array([3, -2, 2j]) # a list
print(f"3d vector as a list {v3_arr}")
```

#### **Transpose of Vectors from row to column and vice versa**

```python
# transpose list

# import numpy as np

v3_1st = np.transpose(v3_1s) # a list
print(f"a 3d vector in a list {v3_1s} and its transpose {v3_1st}")
```

```python
v2_arrt = np.transpose(v2_arr) # transpose from a row array (vector)
print(f"a 2d vector in a row array {v2_arr} and its transpose {v2_arrt}")
```

### **Graph 2d vectors with `mathplotlib`**

```python
#@title 2D vector practice { run: "auto" }
#@markdown Choose if the coordinates are shown:
standard = True #@param {type:"boolean"}

import matplotlib.pyplot as plt
import numpy as np

# set a boolean variable to indicate if the coordinate is shown or not
# standard = False # set False for non-standard vector

def draw(vec_tail, vec_head, color):
    """
    draw vector 
    input: vector(s) for tail and head; color of vectors(s)
    """
    array = np.array([[ vec_tail[0], vec_tail[1], vec_head[0], vec_head[1] ]])
    X, Y, U, V = zip(*array)
    ax.quiver(X, Y, U, V, color = color, angles = "xy", scale_units = "xy", scale = 1)

# main program

# create a figure 
plt.figure (figsize = (5, 5))
ax = plt.gca() # standard axis

# add vectors to the figure
draw([0, 0], [4, 4], "b")
draw([1, 1], [1, 2], "g")

# set limits of viewing window
lim_low = -5
lim_high = 6

ax.set_xlim([lim_low, lim_high])
ax.set_ylim([lim_low, lim_high])

# format the axis
plt.tick_params(
                left = standard, right = standard,
                labelleft = standard, labelbottom = standard,
                bottom = False
                )
plt.figure().patch.set_visible(standard)

if (standard == False):
    ax.axis("off")

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("vector in 2D")

plt.show()
```
