### **Distributive of dot product**

####Given three column vectors with the same dimentionality $n$, then
#### $$\Large \mathbf{a}^\text{T}\mathbf{(b+c)} \,= \, \mathbf{a}^{\text{T}}\mathbf{b} \, +\, \mathbf{a}^{\text{T}}\mathbf{c}  $$

#### or in components

$$\Large  \,\sum_i^n\mathbf{a}_i \left (\mathbf{b}_i+ \mathbf{c}_i \right ) = \,\sum_i^n\mathbf{a}_i\mathbf{b}_i + \,\sum_i^n\mathbf{a}_i\mathbf{c}_i$$

### **TODO2: Writing code for the experiment**
* Define a dimension veriable (soft code)
* Create three radom vectors with the dimension using [`numpy.random.randn(size)`](https://numpy.org/doc/1.16/reference/routines.random.html), where `size` is the number of elements (dimention) in the vector (list)
* Compute the left and right side of the distributive propoerty of dot product 
$$\large \mathbf{a}^\text{T}\mathbf{(b+c)} \,= \, \mathbf{a}^{\text{T}}\mathbf{b} \, +\, \mathbf{a}^{\text{T}}\mathbf{c}  $$
* Compare the results of left and right side

```python
#@title dot product difference { run: "auto" }
dimension = 10000 #@param {type:"integer"}
import numpy as np

# 1. define a dimension variable
# n = 10000

# 2. create 3 vectors with the same dimension
a = np.random.randn(dimension)
b = np.random.randn(dimension)
c = np.random.randn(dimension)

# 3. compute the left and right side of the distributive property of the dot product
left = np.dot(a, b + c) # python does the transpose
right = np.dot(a, b) + np.dot(a, c)

# compare the results
# print(f"Vector a = {a}")
# print(f"Vector b = {b}")
# print(f"Vector c = {c}")
# print(f"\n")
print(f"difference between the left and right sided of the distributive propety")
print(left - right)
```
