**Case 1: dot product is zero** 
  $\mathbf{u} = [-3, 4, 5], \mathbf{v} = [3, 6, -3], \text{s1} = 2$, and $\text{s2} = 3$

* Dot products: $\mathbf{u}\cdot \mathbf{v} = 0$ and $(\text{s1}\mathbf{u})\cdot (\text{s2}\mathbf{v}) = 0$

Discussion: 

1.  Algebraical conclusion: If the dot product of two vectors are zero, the dot product of the scaled vectors will also be zero no matter what scalars we choose because multiplication of multiplicative property of zero: zero multiply any finite number will remain to be zero.  

2. Geometric Interpretation: $\mathbf{u}$ and $\mathbf{v} $ are orthogonal to each other and meet at a right angle.  Multiplication of $\mathbf{u}$ and $\mathbf{v} $ by a scalars, both of these vectors becomes longer or shorter.  But they're still meeting at a right angle and the cosine between these two vectors is still 90 degrees


**Case 2: s1 and s2 both positive** 

 $\mathbf{u} = [-3, 4, 6], \mathbf{v} = [3, 6, -3], \text{s1} = 2$, and $\text{s2} = 3$


* Dot products: $\mathbf{u}\cdot \mathbf{v} = -3$ and $(\text{s1}\mathbf{u})\cdot (\text{s2}\mathbf{v}) = 6 \cdot (-3) = -18$

Discussion: 

1.  Algebraical conclusion: We are taking the dot product of $\mathbf{u}$ and $\mathbf{v} $, which is a single number and scaled it by the product of the two scalars s1 $\cdot$s2 = 6. So the dot product of two scaled factor is -18.

2. Geometric Interpretation: since the dot product of $\mathbf{u}$ and $\mathbf{v} $ are negative, meaning tht they are at an obtuse angle (greater than 90 degree and less than 180 degree.  Since s1$\cdot$s2$>0$ the scaled dot product is positive. The two vectors remins at an obtuse angle after scaling.
