## **Vector Addition and Subtraction: Algebra** 

### $ \color{green}{\textbf{ Dimentionality: }} \textbf{The number of numbers (elements) }$

 **The first and most important** thing to know about doing algebra with vectors is that the dimensionality.  **To add or subtract two vectors,** they must have the same dimentionalities.  You just add or subtract the corresponding elements.

$$
   \begin{bmatrix}
    1 \\
    2 \\
    3 \\
    1+2i\\
    \end{bmatrix} \quad \, +\, \quad\begin{bmatrix}
    -1 \\
    0 \\
    3 \\
    -2 \\
    \end{bmatrix}  = 
    \begin{bmatrix}
    1 + (-1) \\
    2 + 0\\
    3 + 3\\
    1+2i + (-2)\\
    \end{bmatrix}\quad 
    = 
    \begin{bmatrix}
    0 \\
    2\\
    6\\
    -1+2i\\
    \end{bmatrix}
$$

## $$\color{blue}{\textbf{Vector Addition: }} \color{orange}{\textbf{Head}} \textbf{ to } \color{pink}{\textbf{Tail}} $$

```python
#@title Vector Addition $ \begin{bmatrix}1 \\2 \end{bmatrix} \, +\,\begin{bmatrix}2 \\1 \end{bmatrix}\, =\,\begin{bmatrix}3 \\3 \end{bmatrix}$
import numpy  as np
import matplotlib.pyplot as plt

# define two vectors 

v1 = [1, 2]
v2 = [2, 1]
v0 = [0, 0]

def Add_Vec(x, y): 
    """
    Add two vectors
    inputs: Two vectors
    output: coordinates of the two vectors and their addition
    """
    xPlusy = (abs(x[0]+y[0]),abs(x[1]+y[1]))
    array = np.array([[0, 0, x[0], x[1]], 
                      [x[0], x[1], y[0], y[1]], 
                      [0, 0, xPlusy[0], xPlusy[1]]])
    X, Y, U, V = zip(*array)

def draw(vec_tail, vec_head, color):
    """
    Draw vectors
    inputs: vectors for tail and head, and color of the vector
    """
    array = np.array([[vec_tail[0], vec_tail[1], vec_head[0], vec_head[1]]])
    X, Y, U, V = zip(*array)
    plt.quiver(X, Y, U, V,color=color, angles='xy', scale_units='xy',scale=1)

# Create of figure with subplots of 5x5 each in 1 row and 3 column 
plt.figure(figsize=(5 * 3, 5 * 1))  # width by height

# plot the row 1 column 1
plt.subplot(1, 3,1)
ax = plt.gca()

# Add vectors by scall the function draw given the coordinate and tail and head and color
draw([0, 0], v1, "b")
draw([0, 0], v2, "g")

# Set ploting limits
# make sure that it covers the range of all vectors
lim_low = 0
lim_high = 4

plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.title(f"Two Vectors \n v1 and v2")
plt.legend(["v1", "v2"])

# plot the row 1 column 2
plt.subplot(1, 3, 2)
ax = plt.gca()

# Add vectors by scall the function draw given the coordinate and tail and head and color
draw(v0, v1, "b")
draw(v0, v2, "g")

# add two vectors
v1p2 = [v1[0]+v2[0], v1[1]+ v2[1]]
draw(v0, v1p2, "r")

plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.tick_params(left = False, right = True, labelleft = False,
                labelbottom = True, bottom = False)

plt.xlabel('X-axis')
plt.title(f"Addition by Components \n $[v1_x + v2_x,v1_y + v2_y]$")
plt.legend(["v1", "v2", "v1+v2"])

# plot the row 1 column 3
plt.subplot(1, 3, 3)
ax = plt.gca()

# Add vectors by scall the function draw given the coordinate and tail and head and color
draw(v0, v1, "b")

# add two vectors
v1p2 = [v1[0]+v2[0], v1[1]+ v2[1]]

draw(v1, v2, "g")
draw(v0, v1p2, "r")
print(v1p2)
# Set ploting limits
# make sure that it covers the range of all vectors
lim_low = 0
lim_high = 4

plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.tick_params(left = False, right = True, labelleft = False,
                labelbottom = True, bottom = False)

plt.xlabel('X-axis')
plt.title(f"Addition by Tail-to-Head \n from v1 to v2")
plt.legend(["v1", "v2", "v1+v2"])
plt.show()
```

## $$\color{green}{\textbf{Vector Subtraction: }} \color{red}{\textbf{Negative }} \color{blue}{\textbf{Vector Addition}} $$
---

```python
#@title $$\begin{align}\text{Vector Subtraction: }  \begin{bmatrix}1 \\2 \\\end{bmatrix} \, -\,\begin{bmatrix}2 \\1 \end{bmatrix}\, &=\,\begin{bmatrix}-1 \\1 \end{bmatrix} \\ \text{Negative Vector Addition: }  \begin{bmatrix}1 \\2 \\\end{bmatrix} \, +\,\begin{bmatrix}-2 \\-1 \end{bmatrix}\, &=\,\begin{bmatrix}-1 \\1 \end{bmatrix}\end{align}$$
import numpy  as np
import matplotlib.pyplot as plt

# define two vectors 

v1 = [1, 2]
v2 = [2, 1]
v2n = [-2, -1]
v0 = [0, 0]
def Add_Vec(x, y): 
    """
    Add two vectors
    inputs: Two vectors
    output: coordinates of the two vectors and their addition
    """
    xPlusy = (abs(x[0]+y[0]),abs(x[1]+y[1]))
    array = np.array([[0, 0, x[0], x[1]], 
                      [x[0], x[1], y[0], y[1]], 
                      [0, 0, xPlusy[0], xPlusy[1]]])
    X, Y, U, V = zip(*array)

def draw(vec_tail, vec_head, color):
    """
    Draw vectors
    inputs: vectors for tail and head, and color of the vector
    """
    array = np.array([[vec_tail[0], vec_tail[1], vec_head[0], vec_head[1]]])
    X, Y, U, V = zip(*array)
    plt.quiver(X, Y, U, V,color=color, angles='xy', scale_units='xy',scale=1)

# Create of figure with subplots of 5x5 each in 1 row and 3 column 
plt.figure(figsize=(5 * 3, 5 * 1))  # width by height

# plot the vectors in row 1 column 1
plt.subplot(1, 3,1)
ax = plt.gca()

# Add vectors by scall the function draw given the coordinate and tail and head and color
draw([0, 0], v1, "b")
draw([0, 0], v2, "g")
draw(v0, v2n, "orange")

plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.ylabel('Y-axis')
plt.xlabel('X-axis')
plt.title(f"Three Vectors \n v1, v2, and -v2")
plt.legend(["v1", "v2", "-v2"])
plt.grid()

# plot the vectors in row 1 column 2
plt.subplot(1, 3, 2)
ax = plt.gca()

# Add vectors by scall the function draw given the coordinate and tail and head and color
draw(v0, v1, "b")
draw(v0, v2, "g")
draw(v0, v2n, "orange")
# add two vectors
v1m2 = [v1[0]-v2[0], v1[1] - v2[1]]
draw(v0, v1m2, "r")
# Set ploting limits
# make sure that it covers the range of all vectors
plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.tick_params(left = False, right = True, labelleft = False,
                labelbottom = True, bottom = False)

plt.xlabel('X-axis')
plt.title(f"Addition by Components \n $[v1_x + (-v2_x), v1_y + (-v2_y)]$")
plt.legend(["v1", "v2", "-v2", "v1 + (-v2)"])
plt.grid()

# plot the vectors in row 1 column 1
plt.subplot(1, 3, 3)
ax = plt.gca()

# Add vectors by scall the function draw given the coordinate and tail and head and color
draw(v0, v1, "b")

# add two vectors

draw(v0, v2, "g")
draw(v1, v2n, "orange")
draw(v0, v1m2, "r")
# Set ploting limits
# make sure that it covers the range of all vectors
lim_low = -2
lim_high = 3

plt.xlim([lim_low, lim_high])
plt.ylim([lim_low, lim_high])

plt.tick_params(left = False, right = True, labelleft = False,
                labelbottom = True, bottom = False)
plt.xlabel('X-axis')
plt.title(f"Addition by Tail-to-Head \n from v1 to -v2")
plt.legend(["v1", "v2", "-v2", "v1 + (-v2)"])
plt.grid()

plt.show()
```
